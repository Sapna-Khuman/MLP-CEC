```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create an array
x=np.array([0,25])
y=np.array([10,35])
```

```python
#plot the array
plt.plot(x,y)
plt.show()
```

```python
#show the points using marker
plt.plot(x,y,'o')
plt.show()
```

```python
font={"color":"green","size":20}
plt.title("Line Chart",fontdict=font,loc="left")
plt.xlabel("number1")
plt.ylabel("number2")
plt.plot(x1,y1)
plt.grid(axis="y",color="red",linestyle="--",linewidth="1")
plt.grid(axis="x",color="green",linestyle="--",linewidth="1")
```

```python
plt.subplot(1,2,1)
plt.title("p1")
plt.plot(x,y)
plt.subplot(1,2,2)
plt.title("p2")
plt.plot(x1,y1)
plt.suptitle("SuperTitle")
```

```python
#scatter: shows the points 
sizes=np.array([10,50])
plt.scatter(x1,y1,color='red')
plt.scatter(x,y,color='green',s=sizes,alpha=0.8)#Alpha is transparency of points 
plt.show()
```

```python
plt.bar(x1,y1)#bar: plots the graph vertically
plt.show()
```

```python
plt.barh(x1,y1,height=2,color='m')#barh: plots the graph horizontally
plt.show()
```

```python
track_stud=np.array([20,30,40,10,60])
track_name=["Data","IOT","Web","CS","Mobile"]
plt.pie(track_stud,labels=track_name,explode=[0,0.2,0,0,0],shadow=True)
plt.legend(title="Tracks")
```
